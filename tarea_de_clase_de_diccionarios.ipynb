```python
lista_alumnos = []

def agregar(n, e, nt, g, m):
    alumno = {
        "nombre": n,
        "edad": e,
        "nota": nt,
        "grado": g,
        "materias": m
    }
    lista_alumnos.append(alumno)
    print(f"Ya está {n}")

def ver():
    if not lista_alumnos:
        print("No hay nadie en la lista.")
        return

    print(" LISTA ")
    for i, a in enumerate(lista_alumnos):
        m_txt = ", ".join(a['materias'])
        print(f"{i + 1}. {a['nombre']} G:{a['grado']} N:{a['nota']:.1f}")
        print(f"   Temas: {m_txt}")

def promedio():
    if not lista_alumnos:
        print("No hay notas para sacar promedio.")
        return

    s_notas = 0

    for a in lista_alumnos:
        s_notas += a["nota"]

    cantidad = len(lista_alumnos)
    promedio_final = s_notas / cantidad

    print(" PROMEDIO GENERAL ")
    print(f"Total: {cantidad} alumnos.")
    print(f"Promedio: {promedio_final:.2f}")

def borrar():
    ver()
    if not lista_alumnos: return

    i_str = input("Número para borrar: ")

    if i_str.isdigit():
        i = int(i_str) - 1

        if 0 <= i < len(lista_alumnos):
            nombre_borrado = lista_alumnos.pop(i)['nombre']
            print(f"Borrado {nombre_borrado} fuera.")
        else:
            print("Número no existe.")
    else:
        print("Toca un número.")

def cambiar_nota():
    ver()
    if not lista_alumnos: return

    i_str = input("Número para cambiar nota: ")

    if i_str.isdigit():
        i = int(i_str) - 1

        if 0 <= i < len(lista_alumnos):
            a = lista_alumnos[i]

            nueva_nota = input(f"Nota de {a['nombre']} ({a['nota']:.1f}). Nueva nota: ")

            if nueva_nota:
                a['nota'] = float(nueva_nota)
                print(f"Listo Nota de {a['nombre']} cambiada.")
            else:
                print("Todo igual.")
        else:
            print("Número no existe.")
    else:
        print("La nota debe ser un número.")


# Datos de inicio
agregar("Laura M.", 15, 8.5, 10, ["Mate", "Arte"])
agregar("David R.", 16, 7.2, 11, ["Histo", "Edu Fisica"])
agregar("Sofía P.", 14, 9.1, 9, ["Cienc", "Inglés"])
agregar("Manuel G.", 17, 6.8, 12, ["Comp", "Física"])
agregar("Andrea S.", 15, 8.9, 10, ["Españ", "Histo"])

# Menú (Ciclo principal)
while True:
    print("\nMENU DEL GRUPO")
    print("1. Meter alumno")
    print("2. Ver toda la lista")
    print("3. Sacar promedio")
    print("4. Borrar alumno")
    print("5. Cambiar nota")
    print("6. FIN")

    op = input("Elige un número: ")

    if op == '1':
        print(" METER ALUMNO ")
        nombre = input("Nombre: ")
        edad = int(input("Edad: "))
        grado = int(input("Grado: "))
        nota = float(input("Nota: "))
        materias_str = input("Temas (separa con coma): ")
        materias_list = [m.strip() for m in materias_str.split(',')]
        agregar(nombre, edad, nota, grado, materias_list)

    elif op == '2':
        ver()

    elif op == '3':
        promedio()

    elif op == '4':
        borrar()

    elif op == '5':
        cambiar_nota()

    elif op == '6':
        print("Fin")
        break

    else:
        print("Elige 1-6.")6
```
